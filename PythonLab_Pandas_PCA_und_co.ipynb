## Dimensionality Reduction

Principal Component Analysis (PCA): PCA is a popular method for dimensionality reduction that involves transforming a dataset into a new coordinate system that maximizes the variance of the data along the new dimensions. The new dimensions, or principal components, are linear combinations of the original features.

1) Compute the covariance matrix: The covariance matrix represents the relationships between the variables in the dataset. It can be computed using the cov() function from the numpy library in Python.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# load the iris dataset
iris = load_iris()

# center the data
X = iris.data
X_centered = X - np.mean(X, axis=0)

# calculate the covariance matrix
cov_matrix = np.zeros((X.shape[1], X.shape[1]))
for i in range(X.shape[1]):
    for j in range(X.shape[1]):
        cov_matrix[i,j] = np.dot(X_centered[:,i], X_centered[:,j]) / (X.shape[0]-1)

# print the covariance matrix
print(cov_matrix)
```

2) Compute the eigenvectors and eigenvalues of the covariance matrix: The eigenvectors and eigenvalues represent the directions and magnitudes of the variance in the dataset. They can be computed using the eig() function from the numpy library in Python.

```python
# calculate the eigenvectors and eigenvalues of the covariance matrix
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)

# print the eigenvectors and eigenvalues
print("Eigenvalues:\n", eig_vals)
print("\nEigenvectors:\n", eig_vecs)
```

3) Sort the eigenvectors by their corresponding eigenvalues: The eigenvectors with the highest eigenvalues represent the dimensions of the data that capture the most variance.

```python

# sort the eigenvectors by their corresponding eigenvalues
sorted_indices = np.argsort(eig_vals)[::-1]
sorted_eig_vals = eig_vals[sorted_indices]
sorted_eig_vecs = eig_vecs[:, sorted_indices]

# print the sorted eigenvectors and eigenvalues
print("Eigenvalues:\n", sorted_eig_vals)
print("\nEigenvectors:\n", sorted_eig_vecs)
```

4) Select the number of principal components: The number of principal components to retain depends on the desired level of dimensionality reduction and the amount of variance explained by each component.

```python
total_var = np.sum(sorted_eig_vals)

# calculate the explained variance
explained_var = sorted_eig_vals / total_var

# plot the explained variance
plt.plot(np.arange(1, len(explained_var)+1), explained_var, 'bo-')
plt.xlabel('Number of principal components')
plt.ylabel('Explained variance')
plt.show()
```

5) Project the data onto the new coordinate system: The data can be transformed onto the new coordinate system defined by the principal components using the dot product of the standardized data and the eigenvectors.

```python
# select the number of principal components
num_components = 2

# create a projection matrix
projection_matrix = sorted_eig_vecs[:, :num_components]

# project the data onto the new coordinate system
X_pca = np.dot(X_centered, projection_matrix)

# print the shape of the original and projected datasets
print("Original dataset shape: ", X.shape)
print("PCA dataset shape: ", X_pca.shape)
```

6) Analyze the results: The transformed data can be used for visualization or further analysis, such as clustering or classification.

```python
# plot the original and projected data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

colors = ['red', 'green', 'blue']
labels = iris.target_names

for i in range(len(colors)):
    ax1.scatter(X[iris.target==i, 0], X[iris.target==i, 1], alpha=0.5, color=colors[i], label=labels[i])
ax1.set_title("Original Data")
ax1.set_xlabel("Feature 1")
ax1.set_ylabel("Feature 2")
ax1.grid()
ax1.legend()

for i in range(len(colors)):
    ax2.scatter(X_pca[iris.target==i, 0], X_pca[iris.target==i, 1], alpha=0.5, color=colors[i], label=labels[i])
ax2.set_title("PCA Data")
ax2.set_xlabel("Principal Component 1")
ax2.set_ylabel("Principal Component 2")
ax2.grid()
ax2.legend()

plt.show()

# calculate the explained variance ratio
total_var = np.sum(eig_vals)
explained_var_ratio = eig_vals / total_var

# print the explained variance ratio for each principal component
for i in range(num_components):
    print(f"Explained variance ratio of PC{i+1}: {explained_var_ratio[i]:.2f}")
```

PCA function 

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# load the iris dataset
iris = load_iris()

# create a DataFrame with the data
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# center the data
X_centered = df - np.mean(df, axis=0)

# create a PCA object with two components
pca = PCA(n_components=2)

# fit and transform the data using PCA
X_transformed = pca.fit_transform(X_centered)

# create a new DataFrame with the projected data
df_transformed = pd.DataFrame(data=X_transformed, columns=['PC1', 'PC2'])

# concatenate the new DataFrame with the iris target variable
df_transformed = pd.concat([df_transformed, pd.DataFrame(data=iris.target, columns=['target'])], axis=1)

# plot the projected data
fig, ax = plt.subplots()
targets = [0, 1, 2]
colors = ['r', 'g', 'b']
for target, color in zip(targets, colors):
    indicesToKeep = df_transformed['target'] == target
    ax.scatter(df_transformed.loc[indicesToKeep, 'PC1']
               , df_transformed.loc[indicesToKeep, 'PC2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
plt.show()

# compute the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# plot the explained variance ratio
fig, ax = plt.subplots()
ax.bar(np.arange(len(explained_variance_ratio))+1, explained_variance_ratio)
ax.set_xlabel('Principal Component')
ax.set_ylabel('Explained Variance Ratio')
plt.show()
```
